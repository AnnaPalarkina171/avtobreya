```
def make_corpora(text, keys, corpora):
  d = {}
  lemms = list(keys['lemm'])
  poses = list(keys['partlang'])
  count = list(keys['count_word_text'])
  for i,el in enumerate(lemms):
    d[el] = [poses[i], count[i]]

  corpora = corpora.append({'text': text, 'keys': d}, ignore_index=True)
  return corpora

corpora = pd.DataFrame(columns = ['text', 'keys'])
corpora = make_corpora(text_1, keys_1, corpora)
corpora = make_corpora(text_2, keys_2, corpora)
corpora = make_corpora(text_3, keys_3, corpora)
corpora = make_corpora(text_4, keys_4, corpora)
corpora = make_corpora(text_5, keys_5, corpora)
corpora.head()
```

## Применить к этому корпусу 3 метода извлечения ключевых слов на выбор (RAKE, TextRank, tf*idf, OKAPI BM25, ...)

```
!pip install razdel
!pip install pymorphy2

from razdel import tokenize
from razdel import sentenize
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')  
nltk.download('punkt')
from string import punctuation
import pymorphy2

morph = pymorphy2.MorphAnalyzer()
russian_stopwords = stopwords.words('russian')
punctuation = punctuation + '„“'
```

```
def preprocesing(text):
  tokens = list(tokenize(text))
  tokens = [_.text for _ in tokens]
  tokens = [word.lower() for word in tokens if word.lower() not in russian_stopwords]
  tokens = [(word.translate(str.maketrans('', '', punctuation))) for word in tokens]
  lemmas = [morph.parse(str(token))[0].normal_form for token in tokens if token != '']

  return ' '.join(lemmas)

clean_texts = []
for text in list(corpora['text']):
  clean = preprocesing(text)
  clean_texts.append(clean)
corpora['clean_texts'] = clean_texts
```

```
corpora.head()
```

```
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(corpora['clean_texts'])
names = vectorizer.get_feature_names()
data = vectors.todense().tolist()

# Create a dataframe with the results
df = pd.DataFrame(data, columns=names)
df.shape
```

### RAKE

```
!pip install python-rake
import RAKE
```

```
from pymorphy2 import MorphAnalyzer
from pymorphy2.tokenizers import simple_word_tokenize
russian_stopwords.append('это')

# Инициализируем анализатор списком стоп-слов
rake = RAKE.Rake(russian_stopwords)

# эта функция с семинара
m = MorphAnalyzer()
def normalize_text(text):
    lemmas = []
    for t in simple_word_tokenize(text):
        lemmas.append(
            m.parse(t)[0].normal_form
        )
    return ' '.join(lemmas)

rake_keys = []
for text in list(corpora['text']):
  result = rake.run(normalize_text(text), maxWords=3, minFrequency=1)
  rake_keys.append([x for x,y in result])
corpora['rake_keys'] = rake_keys
corpora.head()
```

### TextRank

```
from gensim.summarization import keywords as kw

rank_keys = []
for text in list(corpora['clean_texts']):
  result = kw(normalize_text(text),scores = True)
  rank_keys.append([x for x,y in result])
corpora['rank_keys'] = rank_keys
corpora.head()
```

```
# посмотрим на нграммы эталонных ключевых слов
# отберем уникальные униграммы, 2-граммы и 3-х граммы

def make_gram(i):
  keys = list(corpora['gold_keys'])
  tre_grams = []
  two_grams = []
  uni_grams = []

  for x in keys[i]:
    poses = []
    for y in predictor.predict(x.split()):
      poses.append(y.pos)
    if len(poses) == 1:
      uni_grams.append(poses)
    if len(poses) == 2:
      two_grams.append(poses)
    if len(poses) == 3:
      tre_grams.append(poses)
    

  tre_grams = [x for i, x in enumerate(tre_grams) if i == tre_grams.index(x)]
  two_grams = [x for i, x in enumerate(two_grams) if i == two_grams.index(x)]
  uni_grams = [x for i, x in enumerate(uni_grams) if i == uni_grams.index(x)]
  return tre_grams, two_grams, uni_grams

result = make_gram(0)
first_text = {'unigrams':result[2], 'twograms':result[1], 'tregrams':result[0]}
print(first_text)

result = make_gram(1)
second_text = {'unigrams':result[2], 'twograms':result[1], 'tregrams':result[0]}
print(second_text)

result = make_gram(2)
third_text = {'unigrams':result[2], 'twograms':result[1], 'tregrams':result[0]}
print(third_text)

result = make_gram(3)
forth_text = {'unigrams':result[2], 'twograms':result[1], 'tregrams':result[0]}
print(forth_text)

result = make_gram(4)
fifth_text = {'unigrams':result[2], 'twograms':result[1], 'tregrams':result[0]}
print(fifth_text)
```
